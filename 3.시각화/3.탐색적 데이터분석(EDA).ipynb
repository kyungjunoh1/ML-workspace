#### MinMaxScaler(표준화)
* 모든 특성이 정확하게 0과 1사이에 위치하도록 데이터를 변경한다
* 가장 작은값을 0, 가장 큰값을 1로 두고 결과를 뽑아준다

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
movie = {
    "daum":[2,4,6,8,10],
    "naver":[1,2,3,4,5]
}
mv = pd.DataFrame(movie)
mv
```

```python
min_max_scaler = MinMaxScaler()
#min_max_scaler.fit(mv)
#min_max_scaler.transform(mv)
min_max_mv = min_max_scaler.fit_transform(mv)
```

```python
min_max_mv
```

```python
pd.DataFrame(min_max_mv, columns=['daum','naver'])
```

```python
min_max_scaler = MinMaxScaler()
min_max_df = min_max_scaler.fit_transform(df_sort[['장소부족','비용부족']])
min_max_df
```

```python
df_test = pd.DataFrame(min_max_df, columns=['장소부족','비용부족'])
df_test['분류'] = df_sort['분류'].values
df_test
```

```python
df_sort = df_test[['분류','장소부족','비용부족']]
df_sort.head(2)
```

```python
df_sort['종합점수'] = df_sort["장소부족"] * 0.6 + df_sort['비용부족'] * 0.4
df_sort.head(3)
```

```python
df_sort_top5 = df_sort.sort_values(by="종합점수", ascending=False).head()
df_sort_top5
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,3))
sns.barplot(x="분류", y="종합점수", data=df_sort_top5, ax=ax)
```
