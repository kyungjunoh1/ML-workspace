## 결측데이터 치환

```python
#데이터셋 다시 불러오기
df = pd.read_csv('../../data_set/2.데이터 클린징/bicycle.csv')
df.head()
```

```python
#결측치 확인
df.isnull().sum()
```

```python
#모든 결측치를 0으로 치환. 자기 자신은 변경되지 않습니다
df.fillna(0)
```

```python
#모든 결측치를 이전 값으로 치환
df.ffill()
#이용거리의 2번째 인덱스의 값이 1번째의 값과 동일해 진것을 확인할 수 있다
```

```python
#이용거리 결측치 열을 이용거리 평균으로 치환. 자기자신을 변경

# inplace=True 사용하면 복사본인지 원본인지에 대한 경고 발생
#df['이용거리'].fillna(df['이용거리'].mean(), inplace=True )
df['이용거리'] = df['이용거리'].fillna(df['이용거리'].mean() )
df.head()
#이용거리 2번째 인덱스 값이 평균으로 변경 되었다
```

```python
#결측치 확인
df.isnull().sum()
```

```python
#모든 결측치를 각 열의 평균으로 처리

#데이터안에 문자들이 포함되어 있어 에러 발생
#df.fillna(df.mean(), inplace=True)

#numeric_only : 숫자 대상으로만 연산 가능
df.fillna(df.mean(numeric_only=True), inplace=True)
df.head()
#반납거치대 index3번의 값이 평균으로 변한 내용 확인
```

* 위의 경고는 다음 버전에서는 특정 열만 선택해서 변경하라는 내용이다.
* 이제는 더이상 사용되지 않으며 향후 버전에서는 TypeError이 발생한다는 내용이다
* 경고의 내용은 아래와 같이 해결할 수 있다

```python
columns = ["대여거치대","반납거치대","이용시간"]
#df["대여거치대"].fillna(df["대여거치대"].mean(), inplace=True)
#df["반납거치대"].fillna(df["반납거치대"].mean(), inplace=True)
#df["이용시간"].fillna(df["이용시간"].mean(), inplace=True)
for item in columns:
    df[item] = df[item].fillna(df[item].mean() )
```

```python
#데이터셋 다시 불러오기
df = pd.read_csv('../../data_set/2.데이터 클린징/bicycle.csv')
df.head()
```

```python
# 이용거리를 최빈값으로 치환하세요
print('데이터 개수\n',df['이용거리'].value_counts())
print('-------------------------------')
print('최빈값 : ',df['이용거리'].mode())
df['이용거리'] = df['이용거리'].fillna(df['이용거리'].mode()[0] )
```

```python
#결측 개수 확인. 이용거리 결측값 없어야 된다
print('데이터 행,열 : ',df.shape)
print('---------------------------')
df.isnull().sum()
```
