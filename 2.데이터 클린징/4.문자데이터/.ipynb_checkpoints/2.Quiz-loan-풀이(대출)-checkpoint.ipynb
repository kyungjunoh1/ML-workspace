    - loan_amount_term : 대출기간(개월)
    - credit_history : 신용 기록 지침(0:미충족, 1:충족)
    - property_area : 2:도시, 1:중 도시, 0:농촌
    - loan_status : 대출(1:승인, 0:거절)

```python
#데이터 불러오기
import pandas as pd
df_train = pd.read_csv('../../data_set/2.데이터 클린징/loan-train.csv')
df_train
```

#### 결측치 확인

```python
# 행, 열의 개수 및 결측치 확인
print(df_train.shape)
df_train.isnull().sum()
```

#### 1.결측치 치환
* Gender : Male 또는 Female
* Married : Yes 또는 No
* Dependents : 최빈값과 '3+'는 3으로 변경
* Self_Employed : 이전값
* LoanAmount : 이후값
* Loan_Amount_Term : 평균값
* Credit_History : 1.0 또는 0.0

```python
# Gender 치환 전 값 확인 후 값 치환후 확인
import numpy as np
# Gender 치환 전
print( df_train['Gender'].unique() )
# Gender : Male 또는 Female 치환
df_train['Gender'] = df_train['Gender'].fillna('Male')
np.unique(df_train['Gender'])
# 치환 결과 Female와 Male 두가지만 존재하는 것을 알 수 있다.
```

```python
#Married 치환 전 값 확인 후 값 치환후 확인
# Married 치환 전
print( df_train['Married'].unique() )

# Married : Yes 또는 No 치환 후 결과 확인
df_train['Married'] = df_train['Married'].fillna('Yes')
df_train['Married'].unique()
```

```python
#Dependents 치환 전 값 확인 후 값 치환후 확인
# Dependents 변경 전 확인
print(df_train['Dependents'].unique())

# Dependents : 최빈값과 '3+'는 3으로 변경
# mode()[0] : 해당 결과가 시리즈 값이라 0번째 값으로 가져와야 한다.
df_train['Dependents'] = df_train['Dependents'].fillna( df_train['Dependents'].mode()[0] )
print(df_train['Dependents'].unique())

# Dependents 3+ -> 3으로 변경
df_train = df_train.replace(to_replace = '3+', value='3')
df_train['Dependents'].unique()
```

```python
#Self_Employed 치환 전 값 확인 후 값 치환후 확인
# Self_Employed 변경전 확인
print( df_train['Self_Employed'].unique() )

# Self_Employed 이전 값으로 변경 후 확인
df_train['Self_Employed'] = df_train['Self_Employed'].ffill()
df_train['Self_Employed'].unique()
```

```python
#LoanAmount 치환 전 값 확인 후 값 치환후 확인
# LoanAmount 변경전 확인
print( df_train['LoanAmount'].unique() )

# LoanAmount 이후(다음) 값으로 변경 후 확인
df_train['LoanAmount'] = df_train['LoanAmount'].bfill()
df_train['LoanAmount'].unique()
```

```python
#Loan_Amount_Term 치환 전 값 확인 후 값 치환후 확인
# Loan_Amount_Term 변경전 확인
print( df_train['Loan_Amount_Term'].unique() )

# Loan_Amount_Term 평균 값으로 변경 후 확인
df_train['Loan_Amount_Term'] = df_train['Loan_Amount_Term'].fillna(df_train['Loan_Amount_Term'].mean())
df_train['Loan_Amount_Term'].unique()
```

```python
#Credit_History 치환 전 값 확인 후 값 치환후 확인
# Credit_History 변경전 확인
print( df_train['Credit_History'].unique() )

# Credit_History 1.0 또는 0.0 값으로 변경 후 확인
df_train['Credit_History'] = df_train['Credit_History'].fillna(1.0)
df_train['Credit_History'].unique()
```

```python
# 최종 결과 확인
df_train.isnull().sum()
```

* 결측값 처리 함수화

```python
#데이터 불러오기
import pandas as pd
df_train = pd.read_csv('../../data_set/2.데이터 클린징/loan-train.csv')
df_train.head(3)
```

```python
#모든 컬럼 확인
df_train.columns
```

```python
#모든 결측 값 확인
index = ['Loan_ID', 'Gender', 'Married', 'Dependents', 'Education',
       'Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History', 'Property_Area', 'Loan_Status']
for i in index:
    print( i,' => ',df_train[i].unique() )
```

```python
#결측치 변경하는 함수를 만들어 결측치 치환
def nanChange(col):
    for dic in col:
        for k, v in dic.items():
            if k == 'Self_Employed':
                # method에는 변수를 사용할 수 없어 직접 이름을 써야 한다
                df_train[k] = df_train[k].ffill()
            elif k== 'LoanAmount':
                #method에는 변수를 사용할 수 없어 직접 이름을 써야 한다
                df_train[k] = df_train[k].bfill()
            else:
                df_train[k] = df_train[k].fillna(v)
col = [{'Gender':'Male'},{'Married':'Yes'},{'Dependents':df_train['Dependents'].mode()[0]},{'LoanAmount':'bfill'},
       {'Self_Employed':'pad'},{'Loan_Amount_Term':df_train['Loan_Amount_Term'].mean()},{'Credit_History':1.0}]
df_train.replace(to_replace = '3+', value='3',inplace=True)
nanChange(col)
```

```python
# 최종 결과 확인
df_train.isnull().sum()
```

```python
# 아래 문자열 치환과 비교하기 위해 출력
df_train
```

* 참고
    - 타입 확인 : df.dtypes
    - 문자열 숫자 변환 : df['컬럼'] = pd.to_numeric(df['컬럼'])

```python
#각 컬럼이 가지고 있는 값들의 자료형 확인
df_train.dtypes
```

```python
#위에서 출력된 자료형과 비교
df_train.head(1)
```

```python
#위의 결과에서 숫자형 문자인 컬럼의 유일한 값 출력
df_train['Dependents'].unique()
```

```python
#문자열 숫자를 가지고 있는 값을 숫자로 변경 후 확인
df_train['Dependents'] = df_train['Dependents'].apply(pd.to_numeric)
df_train['Dependents'].unique()
```

```python
#성별(Gender)은 Male : 0, Female : 1로 변경하여 새로운 컬럼 Gender_int 에 저장 후 확인
items = {'Male':0,'Female':1}
df_train['Gender_int'] = df_train['Gender'].map(items)
df_train[['Gender','Gender_int']]
```

```python
#특정 컬럼의 값이 문자열인지 확인
print(type(df_train['Gender'][0]) == str)
```

```python
# 나머지 문자열들은 자기 자신의 컬럼 위치에 숫자로 변환(LabelEncoder사용)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in df_train.columns:
    if type(df_train[column][0]) == str:
        df_train[column] = le.fit_transform(df_train[column])
```

```python
#숫자로 변환 결과 type 확인
df_train.dtypes
```

#### 3. 이상치를 해결하시오
* 최빈값으로 해결 하세요
* 이상치가 없다고 판단되면 변경 안해도 된다.
>
* IQR = Q3(3사분위) - Q1(1사분위)
* 최대 이상치 = Q3(3사분위) + IQR * 1.5(3.0)
* 최저 이상치 = Q1(1사분위) - IQR * 1.5(3.0)

```python
#통계값 확인
df_train.describe()
```

```python
#1사분위와 3사분위의 값을 각각 구해 딕셔너리에 저장 후 출력
# 아이디는 이상치 값을 따로 구할필요가 없다
q1 = {}
q3 = {}
for key in df_train.columns:
    if key != 'Loan_ID':
        q1[key] = df_train[key].describe().loc['25%'] # 1사분위
        q3[key] = df_train[key].describe().loc['75%'] # 3사분위
print(q1)
print(q3)
#위의 내용을 한눈에 보기 위해 데이터 프레임으로 만들었다.
df_test = pd.DataFrame([q1,q3], index=['1사분위','3사분위'])
#df_test
df_test.T
```

```python
#아이디를 제외한 값들의 boxplot 확인
import matplotlib.pyplot as plt
for column in df_train.columns:
    if column != 'Loan_ID':
        #print(column, df_train[column].unique())
        print(column)
        plt.boxplot(df_train[column])
        plt.show()
#결과확인
# Gender성별에서 이상치 가 보이기는 하나 성별은 0,1 둘중 하나이기 때문에 이상치라고 보기 어렵다
#Dependents부양가족 수가 되며 unique()확인해 보면
# Education 교육 여부를 묻는 거고 0,1 두가지이기 때문에 이상치라 보기 어렵다
# Self_Employed 자영업인지 아닌지를 묻고 0,1 이라 이상치라 보기 어렵다
# ApplicantIncome 신청인 소득 역시 수치로 봤을때 문제는 없어 보인다
# CoapplicantIncome 공동 신청자 소득 역시 문제는 없어 보인다
# LoanAmount 대출금 역시 문제 없어 보인다
# Loan_Amount_Term 대출_금액_기간 역시 문제는 없어 보인다
# Credit_History 신용기록도 0,1 두가지라 문제 없어 보인다
```
