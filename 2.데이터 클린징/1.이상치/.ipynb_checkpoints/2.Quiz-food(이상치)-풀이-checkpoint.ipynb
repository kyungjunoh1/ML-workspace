```python
#food.csv의 내용을 불러오세요
import pandas as pd
df = pd.read_csv('../../data_set/2.데이터 클린징/food.csv')
df
```

### 이상치 확인

```python
import matplotlib.pyplot as plt
plt.boxplot(df)
plt.show()
```

* 이상치 구하기
* IQR = Q3(3사분위) - Q1(1사분위)
* 최대 이상치 = Q3(2사분위 끝 지점) + IQR * 1.5(3.0)
* 최저 이상치 = Q1(2사분위 시작 지점) - IQR * 1.5(3.0)

```python
# 각 리스트에 최저, 최고 이상치를 food_a, food_b, food_c 순서대로 넣고 아래쪽 코드에서 이상치 확인을 하자
lowerfence = []
upperfence = []

#반복문 이용하여 각 리스트에 이상치를 저장하는 code 추가
for column in df.columns:
    iqr = df[column].describe().loc['75%'] - df[column].describe().loc['25%']
    lowerfence.append(df[column].describe().loc['25%'] - iqr * 1.5)
    upperfence.append(df[column].describe().loc['75%'] + iqr * 1.5)
    print(column,end=' ')
print('\n------------------------')
print(lowerfence,'최저 이상치')
print(upperfence,' 최고 이상치')
```

```python
# 위에서 만든 이상치 list를 활용하여 각각의 이상치 값을 출력
# 아래의 결과 값 중 처음 값은 food_a가 이상치이며 index 5번째 위치하고 있는 것이다
count = 0
for column in df.columns:
    print(column,'이상치\n---------------\n')
    print(df[(df[column]<lowerfence[count]) | (df[column]> upperfence[count])])
    print('----------------------------------------------\n')
    count = count + 1
```

```python
# food_b의 최저, 최고 이상치 시리즈 값 확인
dd = df['food_b'][(df['food_b']<lowerfence[1]) | (df['food_b']> upperfence[1])]
print(dd.index[0])
dd.loc[ dd.index[0] ]
```

### 1. 이상치 삭제

```python
#반복문을 이용하여 저장되어 있는 이상치를 지우세요.
#삭제 후 확인하면 index 5,8,26인덱스 위치가 삭제된것 확인
# 이상치 행 삭제
count = 0
for column in df.columns:
    s = df[column][(df[column]<lowerfence[count]) | (df[column]> upperfence[count])]
    df.drop(s.index[0], inplace=True)
    count = count + 1
df
```

```python
#위에서 이상치 제거 후 boxplot 확인
import matplotlib.pyplot as plt
plt.boxplot(df)
plt.show()
```

### 2.이상치 중앙값으로 치환 후 확인

```python
# food.csv 파일 다시 불러오기
df = pd.read_csv('../../data_set/2.데이터 클린징/food.csv')
df.shape
```

```python
# df[column] 앞쪽에 column을 꼭 넣어줘야 한다. 그렇지 않으면 a,b,c모두가 30으로 변하게 된다.

count = 0
for column in df.columns:
    mask = (df[column] < lowerfence[count]) | (df[column] > upperfence[count])
    median = df[column].median()
    df.loc[mask, column] = median
    count = count + 1
df
#이전의 이상치 값들이 중앙값으로 치환된다.
#food_a 인덱스 5번째
#food_b 인덱스 26번째
#food_c 인덱스 8번째
```

### 3. 이상치 평균값으로 치환 후 확인

```python
# food.csv 파일 다시 불러오기
df = pd.read_csv('../../data_set/2.데이터 클린징/food.csv')
df.shape
```

```python
# 평균 값이 실수값으로 변경되어 round를 이용하여 소수점수를 버렸다.
#위와 다르게 enumerate 사용
for i, column in enumerate(df.columns):
    mask = (df[column] < lowerfence[i]) | (df[column] > upperfence[i])
    value = round(df[column].mean())
    df.loc[mask, column] = value
df
#food_a 인덱스 5번째
#food_b 인덱스 26번째
#food_c 인덱스 8번째
```

### 4.이상치 최빈값으로 치환 후 확인
* 최빈값이 여러개인 경우 0번째 값으로 사용한다

```python
# food.csv 파일 다시 불러오기
df = pd.read_csv('../../data_set/2.데이터 클린징/food.csv')
df.shape
```

```python
# 각 컬럼별 최빈값 출력
# 최빈값을 확인해 보니 a는 47, b는 20,27,30,40이 동일한 최빈값으로 나온다. c는 20이 나온다.
df = pd.read_csv('../../data_set/2.데이터 클린징/food.csv')
count = 0
for column in df.columns:
    print(column,'최빈값 \n',df[column].mode())
    print('===================')
#food_a 인덱스 5번째
#food_b 인덱스 26번째
#food_c 인덱스 8번째
```

```python
# 각 각의 이상치 값을 최빈값으로 치환
# 최빈값 확인결과 food_b의 마지막이 20으로 잘 처리되었다.
for i, column in enumerate(df.columns):
    mask = (df[column] < lowerfence[i]) | (df[column] > upperfence[i])
    mode_val = df[column].mode()[0]
    df.loc[mask, column] = mode_val
df
```
