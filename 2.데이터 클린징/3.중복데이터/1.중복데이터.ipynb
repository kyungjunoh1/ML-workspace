### 3. 중복 데이터
* 중복 데이터가 존재하면 분석 결과를 왜곡하게 된다.
* 중복 데이터 처리 함수
    - df.duplicated() : 중복된 행(가로)이 있을 경우 True
    - df.drop_duplicates() : 중복되는 행(가로)을 제거
---

```python
import pandas as pd
df = pd.read_csv("../../data_set/2.데이터 클린징/test1.csv")
df
```

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.duplicated().sum()
```
