### 1. 결측치를 해결하시오

* 자전거대여 데이터 셋에서 결측치를 평균으로 변경하시오(단, 평균을 구할 때는 같은 지역(대여소)의 값들의 평균으로 설정해 줘야 합니다)
    * 힌트
        - 특정 대여소 값들 얻어오기 : df[df['대여소번호']==646]

```python
# 결측 데이터 확인하기
df.isnull().sum()
```

```python
# 아래쪽에서 평균값으로 변경후 값 비교하기 위해 해당 코드와 다음 코드를 넣어 놨다.
# 대여소번호가 646과 같은 평균값 확인
df[df['대여소번호']==646].mean(numeric_only=True)
```

```python
# 대여거치대가 null값인 데이터 셋 출력
df[df['대여거치대'].isnull()]
```

```python
#반남거치대가 null 값인 데이터만 확인
df[df['반납거치대'].isnull()]
```

```python
#모든 결측치가 있는 열에 대해서 해당 대여소 번호에 맞는 평균값으로 치환
#소수점 첫번째 자리까지 구한 후 변경해 주세요
# 예) 대여거치대가 NaN값들이 존재 한다면 해당 대여거치대의 대여소 번호중 같은 대여소 번호를 가지고 있는 평균을 구해 해당 대여거치대 값을 처리하세요

import numpy as np
# 결측치가 있는 컬럼만 저장
nan_list = ['대여거치대','반납거치대','이용시간','이용거리']
for number in df['대여소번호'].unique():
    for name in nan_list:
        
        #mean = round(df.loc[df['대여소번호']==number, name].mean(), 1)
        mean = round(df[name][df['대여소번호']==number].mean(),1)
        
        #df[name][df['대여소번호']==number] = df[name][df['대여소번호']==number].fillna(mean)
        #위와 같이 진행하게 되면 복사본을 원본에 설정하는 경고가 뜬다. 그래서 아래처럼 바꾸자
        df.loc[df['대여소번호']==number,name] = df[name][df['대여소번호']==number].fillna(mean)
df.loc[3] #결측값 위치 확인 결과 결측값(반납거치대)이 평균값으로 치환 확인
```

### 2. 이상치를 해결하시오
* 나이에 대애서만 이상치를 처리하세요
* IQR * 1.5 를 할지 IQR * 3을 할지 판단하여 진행 하세요
* 이상치는 중앙값으로 처리하세요. (단, 나이는 소수점이 없으므로 버림, 반올림, 올림 등으로 처리하세요)

```python
# 이상치 확인 결과 나이가 533세라고 하는 것은 이상치가 된다.

# 나이에 관한 통계값 확인
df['나이'].describe()
```

```python
# boxplot을 통한 나이 이상치 확인
import matplotlib.pyplot as plt
plt.boxplot(df['나이'])
plt.show()
```

```python
# 이상치 iqr * 1.5 를 적용하여 확인해 보자
import numpy as np
# 한번에 1사분위와 3사분위 값을 얻어올 수 있다.
q1, q3 = np.percentile(df['나이'], [25,75])
iqr = q3 - q1
upperfence = q3+iqr*1.5
print('iqr*1.5 : ', upperfence)


#나이 이상치 iqr*1.5 결과값들만 확인 df[이상치보다 큰값]
df[df['나이']>upperfence]
#확인결과 iqr*1.5를 적용했더니 연령대가 64세 부터 진행된 것을 확인할 수 있다.
#60대도 자전거를 충분히 탈수 있는 나이이다.
# iqr * 1.5를 하면 안된다는 것이 된다. iqr *3.0으로 다시 계산해 보자
```

```python
# iqr * 3으로 다시 확인
upperfence = q3+iqr*3
print(upperfence)
df[df['나이']>upperfence]
#iqr 3으로 처리하니 어느정도 맞아진다는 것을 알 수 있다.
# 때로는 꼭 iqr 값으로 이상치를 처리할 필요는 없다. 보편적인 값으로 처리해도 된다
```

#### 이상치는 중앙값으로 처리하세요. 

```python
# 이상치를 중앙 값으로 치환

#df[df['나이']>upperfence] = np.median(df['나이'])
df[df['나이']>upperfence] = df['나이'].median()
```

```python
# 나이에 관련된 중복되지 않은 값 출력 후 확인
df['나이'].unique()
```
