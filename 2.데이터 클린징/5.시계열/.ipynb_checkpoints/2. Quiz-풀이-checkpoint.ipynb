# Quiz
### 다음 데이터를 이용하여 데이터를 분석하시오
1. 도시별 가격 평균을 구하시오
2. 각 도시에서 과일별 가격 평균을 구하시오
3. 부산에서 가장 비싼 과일은 몇개인지 구하시오
4. 각 도시별로 가장 비싼 가격과 가장 적은 개수를 구하시오

```python
import pandas as pd
df = pd.DataFrame({
    'city': ['부산', '부산', '부산', '부산', '서울', '서울', '서울'],
    'fruits': ['apple', 'orange', 'banana', 'banana', 'apple', 'apple', 'banana'],
    'price': [100, 200, 250, 300, 150, 200, 400],
    'quantity': [1, 2, 3, 4, 5, 6, 7]
})
df
```

```python
# 1.도시별 가격 평균을 구하시오
df_city = df.groupby( 'city' )
df_city.mean(numeric_only=True)
```

```python
# 2.각 도시에서 과일별 가격 평균을 구하시오
df.groupby(['city','fruits']).mean()
```

```python
# 3.부산에서 가장 비싼 과일은 몇개인지 구하시오
df_city_busan = df_city.get_group('부산')
max_ = df_city_busan['price'].max()
df_city_busan[df_city_busan['price']==max_]
```

```python
# 4.각 도시별로 가장 비싼 가격과 가장 적은 개수를 구하시오
import numpy as np
df_city.agg({'price':'max', 'quantity':'min'})
```
