#### 이용 시간과 이용거리는 대여소명과 반납소명과 일치하는 중위수로 진행 하시오.
* 참고 : df.to_dict('list') DataFrame의 값을 딕셔너리 형태로 바꿔 준다

##### 이용시간 결측치 중앙값으로 치환

```python
# 이용시간 결측치 확인
df[df['이용시간'].isnull()]
```

* 이전과 마찬가지로 결측값이 있는 대여소명과 반납대여소명을 얻어온다

```python
# 이용시간 NaN값 확인
df_time = df[df['이용시간'].isnull()][['대여소명','반납대여소명']]
df_time
```

```python
#연산전 이용시간 중복데이터 개수 얻어오기
df_time.duplicated().sum()
```

```python
# 이용시간 중복데이터 삭제후 저장
df_time.drop_duplicates(inplace=True)
#df = df.drop_duplicates() # 위와 같은 내요이 된다.
df_time
```

```python
# 연산후 이용시간 중복데이터 개수 얻어오기
df_time.duplicated().sum()
```

```python
# DataFrame 딕셔너리로 변경 확인
df_time.to_dict('list')
```

```python
# 위의 내용을 대여소명, 반납대여소명의 각각의 리스트 값으로 변환했다.
time_dic = df_time.to_dict('list')
time_start_list=time_dic.get('대여소명')
time_end_list=time_dic.get('반납대여소명')
```

```python
#장한평역1번출구에서 중량센터의 값들만 가져온다
# 대여소명과 반납대여소명이 같은 정보들을 가져온다.
df[(df['대여소명'] == time_start_list[0]) & (df['반납대여소명'] == time_end_list[0])].head()
```

```python
#이용시간 중앙값 가져오기
df[(df['대여소명'] == time_start_list[0]) & (df['반납대여소명'] == time_end_list[0]) ]['이용시간'].median()
```

```python
# 연산전 이용시간 결측치 확인
df['이용시간'].isnull().sum()
```

```python
# 이용시간 중앙값으로 치환
for i in range( len(time_start_list) ):
    # 이용시간 중앙값 구하기
    time_start = time_start_list[i]
    time_end = time_end_list[i]
    #각 지점의 중앙값을 가져온다
    time_median = df[(df['대여소명'] == time_start) & (df['반납대여소명'] == time_end) ]['이용시간'].median()

    #df.loc[(df['대여소명'] == time_start) & (df['반납대여소명'] == time_end) , '이용시간'] = \
    #        df[(df['대여소명'] == time_start ) & (df['반납대여소명'] == time_end ) ]['이용시간'].fillna( time_median)
    df.loc[ (df['대여소명']==time_start) & (df['반납대여소명'] == time_end ),'이용시간'] = \
            df[(df['대여소명'] == time_start ) & (df['반납대여소명'] == time_end )]['이용시간'].apply( lambda x : time_median if np.isnan(x) else x )
```

```python
# 연산후 이용시간 결측치 확인
df['이용시간'].isnull().sum()
```

##### 이용시간 결측치 중앙값으로 치환

```python
# 이용거리 결측치 확인
df['이용거리'].isnull().sum()
```

```python
# 이용거리 결측값인 대여소명, 반납대여소명 가져오기
df_distance = df[df['이용거리'].isnull()][['대여소명','반납대여소명']]
df_distance
```

```python
# 연산전 이용거리 중복데이터 개수 얻어오기
print('연산전 중복 수 : ',df_distance.duplicated().sum())
# 이용거리 중복데이터 삭제후 저장
df_distance.drop_duplicates(inplace=True)
# 연산후 이용거리 중복데이터 개수 얻어오기
print('연산후 중복 수 : ',df_distance.duplicated().sum())
df_distance
```

```python
# DataFrame 딕셔너리로 변경 확인
# 이용거리. 위의 내용을 각각의 리스트 값으로 변환했다.
distance_dic = df_distance.to_dict('list')
distance_start_list=distance_dic.get('대여소명')
distance_end_list=distance_dic.get('반납대여소명')
```

```python
#장한평역1번출구에서 중량센터의 값들만 가져온다
df[(df['대여소명'] == distance_start_list[0]) & (df['반납대여소명'] == distance_end_list[0]) ].head(2)
```

```python
#중앙값 얻어오기
df[(df['대여소명'] == time_start_list[0]) & (df['반납대여소명'] == time_end_list[0]) ]['이용시간'].median()
```

```python
# 연산전 이용거리 결측치 확인
df['이용거리'].isnull().sum()
```

```python
# 이용시간 중앙값으로 치환
for i in range( len(distance_start_list) ):
    # 이용시간 중앙값 구하기
    start = distance_start_list[i]
    end = distance_end_list[i]
    #각 지점의 중앙값을 가져온다
    d_median = df[(df['대여소명'] == start) & (df['반납대여소명'] == end) ]['이용거리'].median()

    #df.loc[(df['대여소명'] == start) & (df['반납대여소명'] == end) , '이용거리'] = \
    #        df[(df['대여소명'] == start ) & (df['반납대여소명'] == end ) ]['이용거리'].fillna( d_median)
    df.loc[ (df['대여소명']==start) & (df['반납대여소명'] == end ),'이용거리'] = \
            df[(df['대여소명'] == start ) & (df['반납대여소명'] == end )]['이용거리'].apply( lambda x : d_median if np.isnan(x) else x )
```

```python
# 연산후 이용거리 결측치 확인
df['이용거리'].isnull().sum()
```

### 이상치(나이, 이용시간, 이용거리)

* 이용 시간과 거리는 각 대여소번호와 반납대여소번호와 일치하는 중앙값으로 처리하세요(iqr * 1.5)

#### 나이 이상치 중위수로 치환

```python
df['나이'].describe()
#보이는 데이터로 max 533세의 나이는 비정상인 것을 알 수 있다.
```

```python
# 3 이상치를 구해서 결과 확인
import numpy as np

#한번에 1,3사분위 값을 얻어올 수 있다.
q1, q3 = np.percentile(df['나이'],[25,75])
iqr = q3 - q1
upperfence = q3 + iqr * 3
lowerfence = q1 - iqr * 3
print('최고 이상치 * 3 :',upperfence)
print('최저 이상치 * 3 :',lowerfence)
# 확인결과 최고이상치의 연령대는 65세로 나오게 된다.
# 65세까지라고 한다면 지금도 자전거를 탈 수 있는 나이가 된다. 그래서 이상치 값을 * 3으로 진행해야 한다.
# 그리고 최저 이상치는, 나이는 음수가 없다. 음수에 값이 나왔고, 그리고 위의 코드에서 나이의 최저값은 21이기때문에
# 최저 이상치는 따로 구하지 않아도 상관없다.
```

##### 이용시간 대여소번호와 반납대여소번호와 일치하는 중앙값으로 치환

```python
# 1.5 이상치를 구해서 결과 확인
import numpy as np

#한번에 1,3사분위 값을 얻어올 수 있다.
q1, q3 = np.percentile(df[(df['대여소번호']==646) & (df['반납대여소번호']==3)]['이용시간'],[25,75])
iqr = q3 - q1
upperfence = q3 + iqr * 1.5
lowerfence = q1 - iqr * 1.5
print('최고 이상치 * 1.5 :',upperfence)
print('최저 이상치 * 1.5 :',lowerfence)
```

```python
#최저, 최고 이상치 구하는 식 함수화
#대여소번호와 반납대여소번호
def outline( df_re ,rental_num, return_num, name):
    Q1, Q3 = np.percentile(df_re[(df_re['대여소번호']==rental_num) & (df_re['반납대여소번호']==return_num)][name],[25,75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
#모든 대여소번호와 반납대여소번호 가져오기
df_use = df[['대여소번호','반납대여소번호']]
df_use
```

```python
#중복되는 대여소번호와 반납대여소번호 확인
print('연산전 중복데이터 개수 : ',df_use.duplicated().sum())
# df_use.drop_duplicates(inplace=True) 왼쪽의 코드를 사용했더니 경고가 나와서 아래의 내용으로 변경
df_use = df_use.drop_duplicates()
print('연산후 중복데이터 개수 : ',df_use.duplicated().sum())
df_use
```

```python
# 이용시간. 위의 내용을 각각의 리스트 값으로 변환했다.
use_dic = df_use.to_dict('list')
use_start_list=use_dic.get('대여소번호')
use_end_list=use_dic.get('반납대여소번호')
```

```python
#646, 3 의 이상치 값을 확인해서 해당 값들을 변경해 보자
use_start = 646
use_end = 3
lower, upper = outline( df ,use_start, use_end, '이용시간')
print( f"s : {use_start}, e : {use_end} : 이상치 {upper}", end=", 중앙값 : ")
use_median = df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end) ]['이용시간'].median()
print(use_median,"max : ",df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end) ]['이용시간'].max())
#아래의 결과 처럼 이상치를 넘어서는 값들은 중앙값으로 변경해 줄 것이다.
```

```python
#우선 변경하기 전 646, 3의 이상치 보다 큰 데이터를 확인하자.
df.loc[ (df['대여소번호'] == 646) & (df['반납대여소번호'] == 3)&(df['이용시간']>17) ] 
```

```python
#특정 그룹별 값을 변경하고자 할 경우 apply를 이용해야 한다.
df.loc[(df['대여소번호'] == 646) & (df['반납대여소번호'] == 3), '이용시간'] = \
    df[(df['대여소번호'] == 646) & (df['반납대여소번호'] == 3)]['이용시간'].apply(lambda x: 5555 if x > 17 else x )

#변경 후 이전의 이상치 값이 5555로 변경 됐는지 확인하자
df.loc[ (df['대여소번호'] == 646) & (df['반납대여소번호'] == 3)&(df['이용시간']>17) ] 
```

```python
#위의 결과를 토대로 모든 값들을 변경해 주도록 하자
for i in range( len(use_start_list) ):
    # 이용시간 중앙값 구하기
    use_start = use_start_list[i]
    use_end = use_end_list[i]
    lower, upper = outline( df ,use_start, use_end, '이용시간')
    print( f"s : {use_start}, e : {use_end} : 이상치 {upper}", end=", 중앙값 : ")
    use_median = df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end) ]['이용시간'].median()
    print(use_median,"max : ",df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end) ]['이용시간'].max())
    df.loc[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end), '이용시간'] = df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end)]['이용시간'].apply(lambda x: use_median if x > upper else x )
```

```python
# 모든데이터 이상치 변경후 특정 데이터 다시 확인
df[(df['대여소번호']==101) & (df['반납대여소번호']==101)]['이용시간'].describe()
# 위에서 변경하더라도 이상치값을 중앙값으로 변경하다 보면 또 큰숫자들은 이상치가 될 수 있다.
```

##### 이용거리 이상치는 대여소번호와 반납대여소번호와 일치하는 중앙값으로 치환

```python
#위에서 작성한 코드와 동일하며 이용시간 -> 이용거리 로 변경만 하면 끝난다
# 그래서 이런 코드는 함수로 만들어서 사용하면 편리하다
for i in range( len(use_start_list) ):
    # 이용시간 중앙값 구하기
    use_start = use_start_list[i]
    use_end = use_end_list[i]
    lower, upper = outline( df ,use_start, use_end, '이용거리')
    print( f"s : {use_start}, e : {use_end} : 이상치 {upper}", end=", 중앙값 : ")
    use_median = df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end) ]['이용거리'].median()
    print(use_median,"max : ",df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end) ]['이용거리'].max())
    df.loc[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end), '이용거리'] = df[(df['대여소번호'] == use_start) & (df['반납대여소번호'] == use_end)]['이용거리'].apply(lambda x: use_median if x > upper else x )
```
